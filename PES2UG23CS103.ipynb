```python
import numpy as np
from scipy.sparse import spdiags

#1 Determine the length of a matrix or vector
def matrix_length(matrix):
    return len(matrix)

#2 Add B as the fourth row of A to create C
def add_row(A, B):
    return np.insert(A, 3, B, axis=0)

#3 Create D from rows 2, 3, 4 and columns 3, 4 of C
def submatrix(C):
    return C[1:4, 2:4]

#4 Transpose D to create E
def transpose_matrix(D):
    return D.T

#5 Check the size of E
def matrix_size(E):
    return E.shape

#6 Create equally spaced vectors
def create_linspace(start, end, num):
    return np.linspace(start, end, num)

#7 & 8 Find the max and min in each column and row of A
def max_min(A):
    return np.max(A, axis=0), np.min(A, axis=0), np.max(A, axis=1), np.min(A, axis=1)

#9 Calculate mean and sum in each column and row of A
def mean_sum(A):
    return np.mean(A, axis=0), np.mean(A, axis=1), np.sum(A, axis=0), np.sum(A, axis=1)

#10 Create random matrices F and G
def random_matrices(rows, cols):
    return np.random.randint(1, 10, (rows, cols)), np.random.randint(1, 10, (rows, cols))

#11 Perform scalar multiplication, addition, subtraction, and element-wise multiplication on F and G
def matrix_operations(F, G, scalar):
    return scalar * F, F + G, F - G, np.multiply(F, G)

#12 Check size and perform matrix multiplication if compatible
def matrix_multiplication(A, F):
    if A.shape[1] == F.shape[0]:
        return np.dot(A, F)
    return None

#13 Generate special matrices
def special_matrices():
    I = np.eye(3)
    Z = np.zeros((5, 3))
    O = np.ones((4, 2))
    S = np.diag([1, 2, 7])
    R = np.random.randint(1, 10, (6, 6))
    diag_elements = np.diag(R)
    sparse_diag = np.diag([1, 2, 3])
    return I, Z, O, S, diag_elements, sparse_diag

rows_A = int(input("Enter number of rows for matrix A: "))
cols_A = int(input("Enter number of columns for matrix A: "))
A = np.array([list(map(int, input(f"Enter row {i+1} (space-separated values): ").split())) for i in range(rows_A)])

B = np.array(list(map(int, input(f"Enter the row B with {cols_A} elements (space-separated values): ").split())))

scalar = int(input("Enter a scalar value: "))

rows_F = int(input("Enter number of rows for matrix F: "))
cols_F = int(input("Enter number of columns for matrix F: "))
F, G = random_matrices(rows_F, cols_F)

print("Length of matrix A:", matrix_length(A))
C = add_row(A, B)
print("Matrix C after adding B as fourth row:\n", C)
D = submatrix(C)
print("Submatrix D:\n", D)
E = transpose_matrix(D)
print("Transpose E:\n", E)
print("Size of E:", matrix_size(E))
print("Equally spaced vector:", create_linspace(1, 10, 5))
col_max, col_min, row_max, row_min = max_min(A)
print("Column-wise max and min:", col_max, col_min)
print("Row-wise max and min:", row_max, row_min)
col_mean, row_mean, col_sum, row_sum = mean_sum(A)
print("Column mean and sum:", col_mean, col_sum)
print("Row mean and sum:", row_mean, row_sum)
scalar_mult, add_FG, sub_FG, elem_mult_FG = matrix_operations(F, G, scalar)
print("Scalar multiplication:\n", scalar_mult)
print("Matrix addition:\n", add_FG)
print("Matrix subtraction:\n", sub_FG)
print("Element-wise multiplication:\n", elem_mult_FG)
if matrix_multiplication(A, F) is not None:
    print("Matrix multiplication result:\n", matrix_multiplication(A, F))
else:
    print("Matrix multiplication not possible due to incompatible dimensions.")
I, Z, O, S, diag_elements, sparse_diag = special_matrices()
print("Identity matrix:\n", I)
print("Zero matrix:\n", Z)
print("One matrix:\n", O)
print("Diagonal matrix:\n", S)
print("Extracted diagonal elements:", diag_elements)
print("Sparse diagonal matrix:\n", sparse_diag)
```

```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt

# Load the image
image_path="/content/einstein.jpg"
ImJPG = imageio.imread(image_path)

plt.figure(figsize=(6, 6))
plt.imshow(ImJPG, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()

#2 Check image dimensions
print("Image dimensions:", ImJPG.shape)

#3 Check if the image is integer type
print("Image data type:", ImJPG.dtype)

#4 Convert to grayscale (if not already)
if len(ImJPG.shape) == 3:  # If RGB, convert to grayscale
    ImGray = np.mean(ImJPG, axis=2).astype(np.uint8)
else:
    ImGray = ImJPG
plt.imshow(ImGray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

#5 Lighten and Darken Image
ImLight = np.clip(ImGray + 50, 0, 255).astype(np.uint8)
ImDark = np.clip(ImGray - 50, 0, 255).astype(np.uint8)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(ImLight, cmap='gray')
plt.title("Lightened Image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(ImDark, cmap='gray')
plt.title("Darkened Image")
plt.axis("off")
plt.show()

#6 Adjust Contrast
contrast_factor = 1.5  # Increase contrast
ImContrast = np.clip(contrast_factor * (ImGray - 128) + 128, 0, 255).astype(np.uint8)
plt.imshow(ImContrast, cmap='gray')
plt.title("Contrast Adjusted Image")
plt.axis("off")
plt.show()

#7 Crop the image
crop_x, crop_y = 50, 50  # Top-left corner
crop_w, crop_h = 200, 200  # Width and height
ImCrop = ImGray[crop_y:crop_y+crop_h, crop_x:crop_x+crop_w]
plt.imshow(ImCrop, cmap='gray')
plt.title("Cropped Image")
plt.axis("off")
plt.show()

#8 Apply Matrix Transformations
ImFlip = np.flip(ImGray, axis=1)  # Flip horizontally
ImRotated = np.rot90(ImGray)  # Rotate 90 degrees

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(ImFlip, cmap='gray')
plt.title("Flipped Image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(ImRotated, cmap='gray')
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt

# Load image as a matrix
image_path ="/content/HORSE.jpg"
img = imageio.imread(image_path)

# Display Original Image
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

#2 Extracting Image Information
height, width, channels = img.shape
print("Image Dimensions:", height, width, channels)

# Extract RGB channels
red_channel = img[:, :, 0]
green_channel = img[:, :, 1]
blue_channel = img[:, :, 2]

# Display Color Channels
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(red_channel, cmap='Reds')
axs[0].set_title("Red Channel")
axs[1].imshow(green_channel, cmap='Greens')
axs[1].set_title("Green Channel")
axs[2].imshow(blue_channel, cmap='Blues')
axs[2].set_title("Blue Channel")
for ax in axs:
    ax.axis("off")
plt.show()

#3 Grayscale Conversion
gray_img = np.dot(img[..., :3], [0.2989, 0.587, 0.114]).astype(np.uint8)
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

#4 Sepia Filter
sepia_filter = np.array([[0.393, 0.769, 0.189],
                         [0.349, 0.686, 0.168],
                         [0.272, 0.534, 0.131]])
sepia_img = np.dot(img[..., :3], sepia_filter.T).clip(0, 255).astype(np.uint8)
plt.imshow(sepia_img)
plt.title("Sepia Filter")
plt.axis("off")
plt.show()

#5 Red Channel Extraction
red_only = img.copy()
red_only[:, :, 1:] = 0
plt.imshow(red_only)
plt.title("Red Channel Extraction")
plt.axis("off")
plt.show()

#6 Color Permutation (Swap Red & Blue)
swapped_img = img.copy()
swapped_img[:, :, [0, 2]] = swapped_img[:, :, [2, 0]]
plt.imshow(swapped_img)
plt.title("Color Permutation (Red ↔ Blue)")
plt.axis("off")
plt.show()

#7 Remove Green Component
no_green = img.copy()
no_green[:, :, 1] = 0
plt.imshow(no_green)
plt.title("Green Removed")
plt.axis("off")
plt.show()

#8 Color Inversion
inverted_img = 255 - img
plt.imshow(inverted_img)
plt.title("Color Inverted Image")
plt.axis("off")
plt.show()

#9 Saturation Modification (Increase Red & Blue)
saturation_adjust = img.copy().astype(np.float32)
saturation_adjust[:, :, 0] *= 1.2  # Increase Red
saturation_adjust[:, :, 2] *= 1.2  # Increase Blue
saturation_adjust[:, :, 1] *= 0.8  # Decrease Green
saturation_adjust = np.clip(saturation_adjust, 0, 255).astype(np.uint8)
plt.imshow(saturation_adjust)
plt.title("Saturation Modified")
plt.axis("off")
plt.show()

#10 User-defined Color Adjustments
user_matrix = np.array([[1.1, 0.2, 0.1],
                        [0.1, 0.9, 0.2],
                        [0.2, 0.1, 1.0]])
color_adjusted = np.dot(img[..., :3], user_matrix.T).clip(0, 255).astype(np.uint8)
plt.imshow(color_adjusted)
plt.title("User-defined Color Adjustment")
plt.axis("off")
plt.show()

#11 Inverse Transformation
try:
    inverse_matrix = np.linalg.inv(user_matrix)
    restored_img = np.dot(color_adjusted[..., :3], inverse_matrix.T).clip(0, 255).astype(np.uint8)
    plt.imshow(restored_img)
    plt.title("Restored Image (Inverse Transformation)")
    plt.axis("off")
    plt.show()
except np.linalg.LinAlgError:
    print("Inverse matrix does not exist; transformation is not perfectly reversible.")

#12 Gamma Correction
def gamma_correction(image, gamma):
    inv_gamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** inv_gamma * 255 for i in np.arange(0, 256)]).astype(np.uint8)
    return table[image]

brighter_img = gamma_correction(img, 0.5)
darker_img = gamma_correction(img, 2.0)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(brighter_img)
plt.title("Brighter Image (Gamma 0.5)")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(darker_img)
plt.title("Darker Image (Gamma 2.0)")
plt.axis("off")
plt.show()
```

```python
import numpy as np
import scipy.linalg as la
import time

# Define a square matrix A and vector b
A = np.array([[3, 2, -1], [2, -2, 4], [-1, 0.5, -1]])
b = np.array([1, -2, 0])

# Method 1: Using np.linalg.solve
start = time.time()
x1 = np.linalg.solve(A, b)
time_1 = time.time() - start
print("Solution using np.linalg.solve:", x1, "Time taken:", time_1)

# Method 2: LU Decomposition using scipy.linalg.lu
start = time.time()
P, L, U = la.lu(A)
y = la.solve_triangular(L, np.dot(P.T, b), lower=True)
x2 = la.solve_triangular(U, y)
time_2 = time.time() - start
print("Solution using LU Decomposition:", x2, "Time taken:", time_2)

# Method 3: Least Squares Solution
start = time.time()
x3, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)
time_3 = time.time() - start
print("Solution using Least Squares:", x3, "Time taken:", time_3)

# Method 4: Matrix Inverse (only for square matrices)
start = time.time()
A_inv = np.linalg.inv(A)
x4 = np.dot(A_inv, b)
time_4 = time.time() - start
print("Solution using Matrix Inverse:", x4, "Time taken:", time_4)

# Method 5: Row Reduced Echelon Form (RREF)
def rref(A, b):
    Ab = np.hstack([A, b.reshape(-1, 1)])
    m, n = Ab.shape
    for i in range(min(m, n-1)):
        max_row = np.argmax(abs(Ab[i:, i])) + i
        Ab[[i, max_row]] = Ab[[max_row, i]]
        Ab[i] = Ab[i] / Ab[i, i]
        for j in range(i+1, m):
            Ab[j] -= Ab[i] * Ab[j, i]
    for i in range(m-1, -1, -1):
        for j in range(i):
            Ab[j] -= Ab[i] * Ab[j, i]
    return Ab[:, -1]

start = time.time()
x5 = rref(A, b)
time_5 = time.time() - start
print("Solution using RREF:", x5, "Time taken:", time_5)

# Comparing computation times
methods = ["np.linalg.solve", "LU Decomposition", "Least Squares", "Matrix Inverse", "RREF"]
times = [time_1, time_2, time_3, time_4, time_5]
for method, t in zip(methods, times):
    print(f"{method} took {t:.6f} seconds")

# Handling overdetermined system (more equations than variables)
A_over = np.array([[2, 1], [1, -1], [1, 2]])
b_over = np.array([4, 1, 5])
x_over, residuals, rank, s = np.linalg.lstsq(A_over, b_over, rcond=None)
print("Solution for overdetermined system using Least Squares:", x_over)
```
